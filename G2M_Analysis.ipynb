# Merging all four data to produce one data

```python
dataframe1 = pd.merge(cab_data, transaction_data, on='Transaction ID')
dataframe1
```

```python
dataframe2 = pd.merge(dataframe1, customer_data, on='Customer ID')
dataframe2
```

```python
main_data = pd.merge(dataframe2, city_data, on='City')
main_data
```

```python
main_data.isnull().sum()
```

```python
main_data.info()
```

# Changing format of DateTime 

```python
main_data['Date of Travel'] = main_data['Date of Travel'].astype('datetime64[ns]')
```

```python
main_data.info()
```

# Total Profit of each company for the three year period (2016 - 2018)

```python
main_data['Profit'] = main_data['Price Charged'] - main_data['Cost of Trip']
main_data.head()
```

```python
main_data.columns = [col.strip().replace(' ', '_') for col in main_data.columns]
print(main_data.columns)
```

```python
main_data.head()
```

```python
main_data['Year'] = main_data['Date_of_Travel'].dt.year
main_data['Month'] = main_data.Date_of_Travel.dt.month_name()
main_data['Day'] = main_data.Date_of_Travel.dt.day_name()
main_data['Day_of_Week'] = main_data.Date_of_Travel.dt.dayofweek

main_data['Year'] = sorted(main_data['Year'].astype('string'))
```

```python
main_data.info()
```

```python
main_data.isnull().sum()
```

```python
main_data.head()
```

```python
palette = ['#d965a4', '#ffc400']
sns.set_theme()

profit_plot = main_data.groupby(['Company'])['Profit'].sum().reset_index()
profit_plot
```

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(x='Company', y='Profit', data=profit_plot, palette=palette)
ax.bar_label(ax.containers[0])
ax.set(title='Profit for three Years')
```

# Yearly Profit of each Company

```python
yearly_profit = main_data.groupby(['Year', 'Company'])['Profit'].sum().reset_index()
yearly_profit
```

```python
plt.figure(figsize=(14,6))
ax = sns.lineplot(x='Year', y='Profit', data=yearly_profit, hue='Company', palette=palette)
ax.set(title='Yearly Profit of each Company')
plt.show()
```

# Profit made by City

```python
city_profit = main_data.groupby(['City', 'Company'])['Profit'].sum().reset_index()
city_profit
```

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(x='City', y='Profit', data=city_profit, hue='Company', palette=palette)
ax.set(title="City Profit per Company")
plt.xticks(rotation=90)
plt.show()

#city_profit.set_index('City').plot(kind='bar', stacked=True)
```

# Number of rides per year for each company

```python
ride_count = main_data.groupby(['Year', 'Company'])['Price_Charged'].count().reset_index()
ride_count.rename(columns={'Price_Charged': 'No_Of_Rides'}, inplace=True)
ride_count
```

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(x='Year', y='No_Of_Rides', hue='Company', palette=palette, data=ride_count)
ax.set(title="Yearly Number of Rides per Company")
```

# Age group contribution to Profit of each company

```python
main_data.loc[main_data['Age']<=18, 'Age_group'] = 'teenage'
main_data.loc[main_data['Age'].between(19,24), 'Age_group'] = 'yadult'
main_data.loc[main_data['Age'].between(25,39), 'Age_group'] = 'adult'
main_data.loc[main_data['Age']>39, 'Age_group'] = 'older_adult'
```

```python
age_grouping = main_data.groupby([ 'Company', 'Age_group'])[ 'Profit'].sum().reset_index()
age_grouping
```

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(x= age_grouping.Age_group, y=age_grouping.Profit, hue=age_grouping.Company, palette=palette) 
ax.set(title="Age group profit analysis")
plt.xticks(rotation=90)
plt.show()
```

# Monthly Trip Analysis

```python
total_trip_per_month = main_data.groupby(['Month', 'Company'])['Price_Charged'].count().reset_index()
total_trip_per_month.rename(columns={'Price_Charged': 'No_of_Rides'}, inplace=True)
total_trip_per_month
```

```python
plt.figure(figsize=(14, 6))
ax = sns.lineplot(x='Month', y='No_of_Rides', hue='Company', data=total_trip_per_month, palette=palette)
ax.set(title='Monthly number of Rides')
plt.xticks(rotation=90)
plt.show()
```

# DayWise Customer Count

```python
Customer_Day_count = main_data.groupby(['Day', 'Company'])['Customer_ID'].count().reset_index()
Customer_Day_count.rename(columns={'Customer_ID':'Customer_Count'}, inplace=True)
Customer_Day_count
```

```python
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='Day', y='Customer_Count', hue='Company', data=Customer_Day_count, palette=palette)
ax.set(title='Customer Count by Day')
plt.xticks(rotation=90)
plt.show()
```

# Customer by City

```python
customer_by_city = main_data.groupby(['City', 'Company'])['Customer_ID'].count().reset_index()
customer_by_city.rename(columns={'Customer_ID': 'City_Customer_Count'}, inplace=True)
customer_by_city
```

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(x='City', y='City_Customer_Count', hue='Company', data=customer_by_city, palette=palette)
plt.xticks(rotation=90)
plt.show()
```

# EDA Recommendation and Result

The Pink and Yellow cab have been evaluated based on the following analysis:
    1) Total Profit: In the analysis of the total profit for the Yellow and Pink cab, the Yellow cab is seen to have generated
        higher profit for the period of 3years compared to the Pink cab.
    
    2) Yearly Profit: It was seen that the Yellow and Pink cab experienced an increase in their profit from 2016 and 2017
        but at some period say mid 2017 to 2018 there was a drop in profit for both company but the Yellow Cab still had more 
        profit compared to the Pink cab at any point in time during the stated period.
        
    3) Profit made by Cities: This graph shows the city that contributed most to the profit of the company, and it was seen 
        that NewYork City seemed to be the highest contributor to both the Pink and Yellow cabs profit.
        
    4) Number of Rides per Year: This gives us insight to know which cab was more frequently used during the stated period.
        It was seen that from 2016 to 2018 the yellow cab had more rides than the Pink cab there by explaining the reason
        why the profit generated by the Yellow cab is far greater than that of the Pink Cab.
        
    5) Age-group Profit Analysis: In order to fully understand the difference in profit between the Yellow and Pink cab, 
        an insight was further generated to know which age-group makes use of which cab more and it can be seen that all age 
        group makes use of the yellow cab more than the Pink cab, but this also shows that the adult group (25 - 39) are greater 
        users of cabs.
    
    6) Monthly trips: This gives insight to the fact that the Yellow cab has more trips per month than the Pink cab. This 
        could be as a result of location or closing hours of the cab company this could not be shown as there is no data to back this up.
        
    7) Day_wise Customer Count: This gives us insight into the number of customers each cab company had per month and also shows
        that customers tend to use the cabs more during the weekend (Friday, Saturday and Sunday). In this weekend days it is seen 
        that the Yellow cab has more customers than the Pink Cab.
    
    8) Customer by City: This graph gives insight to the customer base of each cab company to know which city uses which cab more.
        It shows that the Yellow cab has more customers in NewYork, Chicago and Washington DC and Boston while the Pink
        cab has more of its customers in Los Angeles, New York, and San Diego. It would also be nice to note that in the countries
        where Pink cab has most of it's customer, the Yellow cab still has more customers than them.
        
Based on the above insights, I would recommend investment on the Yellow Cab Company.
